```python
# URL del archivo CSV
url = "https://ucdp.uu.se/downloads/ged/ged221-csv.zip"

# Este metodo descarga, extrae el archivo ZIP y lee el archivo csv
def download_and_extract(url, extract_to='data'):
    # Descargar el archivo
    response = requests.get(url)
    zip_path = extract_to + '.zip'
    
    with open(zip_path, 'wb') as file:
        file.write(response.content)

    # Extraer el archivo ZIP
    from zipfile import ZipFile
    with ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

    # Leer el archivo CSV
    csv_path = extract_to + '/GEDEvent_v22_1.csv'
    df = pd.read_csv(csv_path, encoding='latin1')
    return df

# Ejemplo de uso
df = download_and_extract(url)
```

```python
# Ejemplo de Limpieza de Datos

# Eliminar duplicados
def transform_data(df):
    df = df.drop_duplicates()

    return df
    
# Eliminar filas con valores nulos en columnas críticas
columns_to_check = ['year', 'active_year', 'country_id', 'type_of_violence']
df = df.dropna(subset=columns_to_check) 
```

```python
# Filtrar 5 conflictos por pais y eliminar filas no deseadas

def filter_conflicts_by_country(df, n=5):
    # Filtrar para obtener solo los primeros 'n' conflictos por país
    filtered_df = df.groupby('country').head(n)
    return filtered_df
```

```python
# Uso
data = download_and_extract(url)
df = transform_data(data)

print(df.columns)
    
# Mostrar las primeras filas del DataFrame transformado
print(df.head())

# Guardar el DataFrame transformado en un nuevo archivo CSV
df.to_csv('data/GEDEvent_v22_1_transformed.csv', index=False)
```

```python
from datetime import datetime

# Metodo para limpiar fechas
def limpiar_fechas(fecha):
    
    fecha = fecha.split()[0]
    fecha = datetime.strptime(fecha, "%Y-%m-%d")
    
    return fecha

df["date_start"] = df["date_start"].apply(lambda x : limpiar_fechas(x))
df["date_end"] = df["date_end"].apply(lambda x : limpiar_fechas(x))
```
